##### Data Visualization

```python
df['children']=df['children'].astype('object')
```

```python
# Plotting numerical columns

num_cols = df.select_dtypes(exclude = 'object')
num_cols.columns
```

```python
# Plotting categorical columns

cat_cols = df.select_dtypes(include = 'object')
cat_cols.columns
```

```python
plt.figure(figsize = (15,6))
f_num = cat_cols.columns[:]

for i in enumerate(f_num):
    f = plt.figure(figsize=(15,5))
    ax = f.add_subplot(121)
    sns.countplot(cat_cols[i[1]])
    plt.show()
```

```python
sns.barplot('region', 'expenses', hue='sex' ,data=df);
```

```python
sns.barplot('region', 'expenses', hue='smoker' ,data=df)
```

```python
fig = plt.figure(figsize=(10,5))
sns.scatterplot('bmi','expenses',hue='smoker', data=df)
plt.show()
```

```python
plt.figure(figsize = (15,6))
f_num = num_cols.columns[:]

for i in enumerate(f_num):
    f = plt.figure(figsize=(15,5))
    ax = f.add_subplot(121)
    sns.boxplot(num_cols[i[1]], color= 'Blue')

    ax1 = f.add_subplot(122)
    sns.distplot(num_cols[i[1]], rug = True,  color= 'Blue')
    plt.show()
```
